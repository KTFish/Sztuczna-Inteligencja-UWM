```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
def plot_model_performance(df, name):
    fig, ax = plt.subplots(2, 1)

    # Set the colors and labels
    line_colors = ['black', 'red']
    line_labels = ['Training Loss', 'Validation Loss', 'Training Accuracy', 'Validation Accuracy']

    # Plot the lines
    ax[0].plot(df['loss'], color=line_colors[0], label=line_labels[0])
    ax[0].plot(df['val_loss'], color=line_colors[1], label=line_labels[1])
    ax[0].legend()

    ax[1].plot(df['accuracy'], color=line_colors[0], label=line_labels[2])
    ax[1].plot(df['val_accuracy'], color=line_colors[1], label=line_labels[3])
    ax[1].legend()

    # Add titles and axis labels
    ax[0].set_title('Loss')
    ax[1].set_title('Accuracy')
    fig.suptitle(f'{name} Model Performance')
    ax[1].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[1].set_ylabel('Accuracy')
    
    ax[0].grid(linewidth=0.3)
    ax[1].grid(linewidth=0.3)
    plt.gca().set_ylim(0, 1)

    plt.tight_layout()
    plt.show()
```

### Naive Model

```python
naive = pd.read_csv('./results/fashion/naive_model.csv').drop(['Unnamed: 0'], axis=1)
plot_model_performance(naive, 'Naive')
```

### Over trained model

```python
df = pd.read_csv('./results/fashion/better_model.csv').drop(['Unnamed: 0'], axis=1)
plot_model_performance(df, 'Overtrained')
```

### Final Model

```python
df = pd.read_csv('./results/fashion/final.csv').drop(['Unnamed: 0'], axis=1)
plot_model_performance(df, 'Final Model')
```
